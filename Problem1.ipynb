```python
# sort elements based on second position

n = int(input ("Enter number of elements:  "))

l = list(map(str, input().split(" ")[:n]))

for i in range(len(l)):
    for j in range(len(l)):
        #print(l[i][-2], l[j][-2])

        if l[j][-2] > l[i][-2]:
            temp = l[i]
            l[i] = l[j]
            l[j] = temp
print(l)
```
